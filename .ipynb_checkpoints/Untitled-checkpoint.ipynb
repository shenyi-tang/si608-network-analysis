```python

edges_1994 = pref[['origin', 'destination', 'year_1994']].rename(columns={'year_1994': 'travel_time'})

# 去掉缺失或无意义的通行时间
edges_1994 = edges_1994.dropna(subset=['travel_time'])
edges_1994 = edges_1994[edges_1994['travel_time'] > 0]

# -----------------------------
# 3. 合并坐标信息
# -----------------------------
edges_1994 = (
    edges_1994
    .merge(cityinfo[['dzcode', 'coord_long', 'coord_lat', 'cityname_eng']], 
           left_on='origin', right_on='dzcode', how='left')
    .rename(columns={'coord_long': 'x1', 'coord_lat': 'y1', 'cityname_eng': 'city1'})
    .merge(cityinfo[['dzcode', 'coord_long', 'coord_lat', 'cityname_eng']], 
           left_on='destination', right_on='dzcode', how='left')
    .rename(columns={'coord_long': 'x2', 'coord_lat': 'y2', 'cityname_eng': 'city2'})
)

edges_1994.dropna(subset=['x1', 'y1', 'x2', 'y2'], inplace=True)

# -----------------------------
# 4. 构建 NetworkX 图
# -----------------------------
G = nx.from_pandas_edgelist(
    edges_1994,
    source='city1',
    target='city2',
    edge_attr='travel_time',
    create_using=nx.Graph()
)

# 节点坐标映射
pos = {row.cityname_eng: (row.coord_long, row.coord_lat) for _, row in cityinfo.iterrows()}

# -----------------------------
# 5. 设置颜色映射：颜色 = 通行时间长短
# -----------------------------
travel_times = [d['travel_time'] for _, _, d in G.edges(data=True)]
vmin, vmax = np.percentile(travel_times, [5, 95])  # 去掉极端值
norm = plt.Normalize(vmin=vmin, vmax=vmax)
cmap = plt.cm.coolwarm

# -----------------------------
# 6. 可视化网络
# -----------------------------
# -----------------------------
# 6. 可视化网络（修正版）
# -----------------------------
fig, ax = plt.subplots(figsize=(10, 8))

ax.set_title("China Prefecture-level Road Network (1994)", fontsize=14)

# 画边
for u, v, data in G.edges(data=True):
    x1, y1 = pos[u]
    x2, y2 = pos[v]
    color = cmap(norm(data['travel_time']))
    ax.plot([x1, x2], [y1, y2],
            color=color,
            linewidth=0.7 + 2.5 * (1 - (data['travel_time'] - vmin) / (vmax - vmin)),
            alpha=0.7)

# 画节点
ax.scatter(cityinfo.coord_long, cityinfo.coord_lat, s=10, c='black', zorder=3)

# 添加 colorbar —— 必须加 ax 参数！
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, shrink=0.6)
cbar.set_label('Travel Time (hours)')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.grid(False)

plt.tight_layout()
plt.show()
```

```python
# Step 1: 找每个 segment 的起点与终点坐标
start_points = pixel.groupby('seg_id').first().reset_index()[['seg_id', 'long', 'lat']]
end_points   = pixel.groupby('seg_id').last().reset_index()[['seg_id', 'long', 'lat']]

# Step 2: 建立 KDTree 以快速查找最近城市
city_coords = np.array(list(zip(cityinfo.coord_long, cityinfo.coord_lat)))
city_tree = cKDTree(city_coords)

def nearest_city(lon, lat):
    dist, idx = city_tree.query([lon, lat])
    return cityinfo.iloc[idx].cityname_eng

# Step 3: 匹配城市
start_points['city_start'] = start_points.apply(lambda r: nearest_city(r['long'], r['lat']), axis=1)
end_points['city_end'] = end_points.apply(lambda r: nearest_city(r['long'], r['lat']), axis=1)

# Step 4: 合并为边表
edges = start_points.merge(end_points, on='seg_id', suffixes=('_start', '_end'))
edges = edges[edges['city_start'] != edges['city_end']]  # 去掉自环

edges.head()

def haversine(lon1, lat1, lon2, lat2):
    R = 6371  # 地球半径 km
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

edges['distance_km'] = edges.apply(lambda r: haversine(r['long_start'], r['lat_start'], 
                                                       r['long_end'], r['lat_end']), axis=1)
G = nx.Graph()

# 添加节点
for _, row in cityinfo.iterrows():
    G.add_node(row['cityname_eng'], pos=(row['coord_long'], row['coord_lat']))

# 添加边（带距离权重）
for _, row in edges.iterrows():
    G.add_edge(row['city_start'], row['city_end'], weight=row['distance_km'])

plt.figure(figsize=(10, 8))
pos = nx.get_node_attributes(G, 'pos')

# 绘制边
nx.draw_networkx_edges(G, pos, width=0.4, alpha=0.6, edge_color='gray')

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_size=20, node_color='red')

# 可选：只标出主要城市
main_cities = ['Beijing', 'Shanghai', 'Wuhan', 'Guangzhou', 'Chengdu']
nx.draw_networkx_labels(G, pos, labels={c: c for c in main_cities}, font_size=8, font_color='darkred')

plt.title("China City-level Railway Network (2017)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
import networkx as nx
import matplotlib.pyplot as plt

# === Step 1: 读取数据 ===
cityinfo = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

# === Step 2: 建立城市坐标索引树 ===
city_coords = np.array(list(zip(cityinfo.coord_long, cityinfo.coord_lat)))
city_tree = cKDTree(city_coords)

def nearest_city(lon, lat):
    dist, idx = city_tree.query([lon, lat])
    return cityinfo.iloc[idx].cityname_eng

# === Step 3: 为每个像素匹配最近城市 ===
pixel['nearest_city'] = pixel.apply(lambda r: nearest_city(r['long'], r['lat']), axis=1)

# === Step 4: 按 seg_id 分组，追踪城市变化 ===
edges_list = []

for seg, group in pixel.groupby('seg_id'):
    cities_in_path = group['nearest_city'].tolist()
    # 去除重复连续城市
    unique_cities = [cities_in_path[i] for i in range(1, len(cities_in_path)) 
                     if cities_in_path[i] != cities_in_path[i-1]]
    # 相邻城市成边
    for i in range(len(unique_cities)-1):
        a, b = unique_cities[i], unique_cities[i+1]
        if a != b:
            edges_list.append((a, b))

# === Step 5: 聚合边（去重、统计出现次数）===
edges_df = pd.DataFrame(edges_list, columns=['city_start', 'city_end'])
edges_df['pair'] = edges_df.apply(lambda x: tuple(sorted([x['city_start'], x['city_end']])), axis=1)
edges_df = edges_df.groupby('pair').size().reset_index(name='count')
edges_df[['city_start', 'city_end']] = pd.DataFrame(edges_df['pair'].tolist(), index=edges_df.index)
edges_df = edges_df.drop(columns='pair')

print(edges_df.head())

G_real = nx.Graph()

# 添加城市节点
for _, row in cityinfo.iterrows():
    G_real.add_node(row['cityname_eng'], pos=(row['coord_long'], row['coord_lat']))

# 添加铁路边（真实连通）
for _, row in edges_df.iterrows():
    G_real.add_edge(row['city_start'], row['city_end'], weight=row['count'])

plt.figure(figsize=(10, 8))
pos = nx.get_node_attributes(G_real, 'pos')

# 绘制边
weights = [0.2 + 2*np.log1p(G_real[u][v]['weight']) for u,v in G_real.edges()]  # 边宽随出现次数加粗
nx.draw_networkx_edges(G_real, pos, width=weights, alpha=0.6, edge_color='gray')

# 绘制节点
nx.draw_networkx_nodes(G_real, pos, node_size=30, node_color='red')

# 标注主要城市
main_cities = ['Beijing', 'Shanghai', 'Wuhan', 'Guangzhou', 'Chengdu']
nx.draw_networkx_labels(G_real, pos, labels={c:c for c in main_cities}, font_size=8, font_color='darkred')

plt.title("China City-level Railway Network (Real Connectivity, 2017)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
import networkx as nx
import matplotlib.pyplot as plt

# === Step 1: 读入数据 ===
cityinfo = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

# === Step 2: 为每个 pixel 匹配最近城市 ===
city_coords = np.array(list(zip(cityinfo.coord_long, cityinfo.coord_lat)))
city_tree = cKDTree(city_coords)

def nearest_city(lon, lat):
    dist, idx = city_tree.query([lon, lat])
    return cityinfo.iloc[idx].cityname_eng, dist

pixel[['nearest_city', 'dist_km']] = pixel.apply(lambda r: pd.Series(nearest_city(r['long'], r['lat'])), axis=1)

# 过滤掉距离城市太远的pixel（比如 > 30km），避免误连边
pixel = pixel[pixel['dist_km'] < 0.3]  # 约30km以内

edges_list = []

for seg, group in pixel.groupby('seg_id'):
    cities_on_seg = group['nearest_city'].unique()
    if len(cities_on_seg) >= 2:
        for i in range(len(cities_on_seg)-1):
            for j in range(i+1, len(cities_on_seg)):
                edges_list.append((cities_on_seg[i], cities_on_seg[j]))

edges_df = pd.DataFrame(edges_list, columns=['city_start', 'city_end'])
edges_df['pair'] = edges_df.apply(lambda x: tuple(sorted([x.city_start, x.city_end])), axis=1)
edges_df = edges_df.groupby('pair').size().reset_index(name='count')
edges_df[['city_start', 'city_end']] = pd.DataFrame(edges_df['pair'].tolist(), index=edges_df.index)
edges_df = edges_df.drop(columns='pair')

G = nx.Graph()

# 添加节点
for _, row in cityinfo.iterrows():
    G.add_node(row['cityname_eng'], pos=(row['coord_long'], row['coord_lat']))

# 添加边
for _, row in edges_df.iterrows():
    G.add_edge(row['city_start'], row['city_end'], weight=row['count'])

plt.figure(figsize=(10, 8))
pos = nx.get_node_attributes(G, 'pos')

# 绘制边（加权宽度）
weights = [0.5 + 1.5*np.log1p(G[u][v]['weight']) for u,v in G.edges()]
nx.draw_networkx_edges(G, pos, width=weights, alpha=0.4, edge_color='gray')

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_size=25, node_color='red')

# 标注主要城市
main_cities = ['Beijing', 'Shanghai', 'Wuhan', 'Guangzhou', 'Chengdu']
nx.draw_networkx_labels(G, pos, labels={c:c for c in main_cities}, font_size=8, font_color='darkred')

plt.title("China City Railway Network (True Pixel Connectivity, 2017)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
import networkx as nx
import matplotlib.pyplot as plt

# === Step 1: 读取数据 ===
cityinfo = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

# === Step 2: 匹配每个像素的最近城市 ===
city_coords = np.array(list(zip(cityinfo.coord_long, cityinfo.coord_lat)))
city_tree = cKDTree(city_coords)

def nearest_city(lon, lat):
    dist, idx = city_tree.query([lon, lat])
    return cityinfo.iloc[idx].cityname_eng, dist

pixel[['nearest_city', 'dist']] = pixel.apply(lambda r: pd.Series(nearest_city(r['long'], r['lat'])), axis=1)
pixel = pixel[pixel['dist'] < 0.4]  # 40km 内的像素才认为属于城市范围

# === Step 3: 基于铁路轨迹顺序构建城市边 ===
edges_list = []

for seg, group in pixel.groupby('seg_id'):
    # 轨迹按 index 或 pos_y 排序
    group = group.sort_values(by=['index'])
    # 取唯一连续城市序列
    city_seq = group['nearest_city'].tolist()
    city_seq = [city_seq[i] for i in range(1, len(city_seq)) if city_seq[i] != city_seq[i-1]]
    
    # 依次连边（真实顺序）
    for i in range(len(city_seq) - 1):
        a, b = city_seq[i], city_seq[i+1]
        if a != b:
            edges_list.append((a, b))

edges_df = pd.DataFrame(edges_list, columns=['city_start', 'city_end'])
edges_df['pair'] = edges_df.apply(lambda x: tuple(sorted([x.city_start, x.city_end])), axis=1)
edges_df = edges_df.groupby('pair').size().reset_index(name='count')
edges_df[['city_start', 'city_end']] = pd.DataFrame(edges_df['pair'].tolist(), index=edges_df.index)
edges_df = edges_df.drop(columns='pair')

print(edges_df.head())

G = nx.Graph()

# 添加城市节点
for _, row in cityinfo.iterrows():
    G.add_node(row['cityname_eng'], pos=(row['coord_long'], row['coord_lat']))

# 添加真实铁路边
for _, row in edges_df.iterrows():
    G.add_edge(row['city_start'], row['city_end'], weight=row['count'])

plt.figure(figsize=(10, 8))

# 绘制原始铁路像素轨迹（灰线背景）
for seg, group in pixel.groupby('seg_id'):
    plt.plot(group['long'], group['lat'], color='black', linewidth=0.3, alpha=0.3)

# 绘制城市铁路网络（红线）
pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx_edges(G, pos, width=1, alpha=0.8, edge_color='red')
nx.draw_networkx_nodes(G, pos, node_size=20, node_color='orange', edgecolors='k', linewidths=0.2)

# 标注主要城市
main_cities = ['Beijing', 'Shanghai', 'Wuhan', 'Chengdu', 'Guangzhou']
nx.draw_networkx_labels(G, pos, labels={c: c for c in main_cities}, font_size=8, font_color='darkred')

plt.title("China Railway Network (City Nodes from Pixel Trajectories, 2017)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(alpha=0.2)
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
import networkx as nx
import matplotlib.pyplot as plt

# === Step 1: 读数据 ===
cityinfo = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

# === Step 2: 建立 KDTree 方便查找最近城市 ===
city_coords = np.array(list(zip(cityinfo.coord_long, cityinfo.coord_lat)))
city_tree = cKDTree(city_coords)

def nearest_city(lon, lat):
    dist, idx = city_tree.query([lon, lat])
    return cityinfo.iloc[idx].cityname_eng, dist

# === Step 3: 遍历每条铁路 ===
edges = []
for seg_id, group in pixel.groupby('seg_id'):
    group = group.sort_values(by='index')  # 按轨迹顺序
    start = group.iloc[0]
    end = group.iloc[-1]
    
    city_start, dist1 = nearest_city(start['long'], start['lat'])
    city_end, dist2 = nearest_city(end['long'], end['lat'])
    
    if city_start != city_end and dist1 < 0.4 and dist2 < 0.4:  # 确保端点都在城市附近（约40km）
        edges.append((city_start, city_end))

edges_df = pd.DataFrame(edges, columns=['city_start', 'city_end'])
edges_df['pair'] = edges_df.apply(lambda x: tuple(sorted([x.city_start, x.city_end])), axis=1)
edges_df = edges_df.groupby('pair').size().reset_index(name='count')
edges_df[['city_start', 'city_end']] = pd.DataFrame(edges_df['pair'].tolist(), index=edges_df.index)
edges_df = edges_df.drop(columns='pair')

print(edges_df.head())

G = nx.Graph()

# 添加城市节点
for _, row in cityinfo.iterrows():
    G.add_node(row['cityname_eng'], pos=(row['coord_long'], row['coord_lat']))

# 添加铁路边
for _, row in edges_df.iterrows():
    G.add_edge(row['city_start'], row['city_end'], weight=row['count'])

plt.figure(figsize=(10, 8))

# 1️⃣ 绘制铁路像素轨迹背景
for seg, group in pixel.groupby('seg_id'):
    plt.plot(group['long'], group['lat'], color='black', linewidth=0.3, alpha=0.4)

# 2️⃣ 绘制城市铁路 network
pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx_edges(G, pos, edge_color='red', width=1.2, alpha=0.8)
nx.draw_networkx_nodes(G, pos, node_size=25, node_color='orange', edgecolors='k', linewidths=0.2)

# 3️⃣ 标注主要城市
main_cities = ['Beijing', 'Shanghai', 'Wuhan', 'Guangzhou', 'Chengdu']
nx.draw_networkx_labels(G, pos, labels={c: c for c in main_cities}, font_size=8, font_color='darkred')

plt.title("China City Railway Network (Based on Segment Endpoints, 2017)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(alpha=0.2)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from shapely.geometry import Point, LineString
from sklearn.cluster import DBSCAN

# Step 1: 读取数据


# 转为 GeoDataFrame
gdf_cities = gpd.GeoDataFrame(
    cityinfo, geometry=gpd.points_from_xy(cityinfo.coord_long, cityinfo.coord_lat), crs="EPSG:4326"
)
gdf_pixels = gpd.GeoDataFrame(
    pixel, geometry=gpd.points_from_xy(pixel.long, pixel.lat), crs="EPSG:4326"
)

# 转投影到平面坐标（便于距离计算）
gdf_cities = gdf_cities.to_crs(epsg=3857)
gdf_pixels = gdf_pixels.to_crs(epsg=3857)

# Step 2: 聚类铁路点（每条铁路一类）
coords = np.array(list(zip(gdf_pixels.geometry.x, gdf_pixels.geometry.y)))
db = DBSCAN(eps=8000, min_samples=8).fit(coords)  # eps单位：米，调大可连通更长轨道
gdf_pixels['cluster'] = db.labels_

rail_lines = []
for cid, group in gdf_pixels.groupby('cluster'):
    if cid == -1:  # 忽略噪声点
        continue
    group_sorted = group.sort_values(by=['lat', 'long'])
    line = LineString(group_sorted.geometry.tolist())
    rail_lines.append({'cluster': cid, 'geometry': line})

gdf_lines = gpd.GeoDataFrame(rail_lines, crs=gdf_pixels.crs)

# Step 3: 找出城市关联的铁路
city_to_line = []
for idx, city in gdf_cities.iterrows():
    distances = gdf_lines.distance(city.geometry)
    nearest = distances.idxmin()
    if distances[nearest] < 15000:  # 若城市距某条铁路<15km，认为连接
        city_to_line.append((city['cityname_eng'], nearest))

# Step 4: 构建城市网络
G = nx.Graph()
for city in gdf_cities['cityname_eng']:
    G.add_node(city)

for line_id in set([lid for _, lid in city_to_line]):
    cities_on_line = [c for c, lid in city_to_line if lid == line_id]
    for i in range(len(cities_on_line)):
        for j in range(i + 1, len(cities_on_line)):
            G.add_edge(cities_on_line[i], cities_on_line[j])

# Step 5: 可视化
fig, ax = plt.subplots(figsize=(12, 10))
gdf_lines.plot(ax=ax, color='black', linewidth=1, alpha=0.6, label='Railways')
gdf_cities.plot(ax=ax, color='red', markersize=20, label='Cities')

# 可选：绘制边
for u, v in G.edges():
    city_u = gdf_cities[gdf_cities['cityname_eng'] == u].geometry.values[0]
    city_v = gdf_cities[gdf_cities['cityname_eng'] == v].geometry.values[0]
    line = LineString([city_u, city_v])
    gpd.GeoSeries([line]).plot(ax=ax, color='red', linewidth=0.5, alpha=0.5)

plt.title("China Railway City Network (based on pixel proximity)")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```

```python
import pandas as pd
import networkx as nx
from scipy.spatial import cKDTree

city = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()
# Step 2: 为每条铁路构建空间索引（更快查找城市附近的像素）
results = []

for seg, group in pixel.groupby('seg_id'):
    # 用 pos_x, pos_y 构建 KDTree（快速查找）
    tree = cKDTree(group[['pos_x', 'pos_y']].values)
    
    for _, city_row in city.iterrows():
        dist, idx = tree.query([city_row['pos_x'], city_row['pos_y']])
        if dist < 20:  # 可调：如果像素差小于20格，视为铁路经过城市
            results.append((city_row['cityname_eng'], seg))

# Step 3: 将结果转为 DataFrame
df_link = pd.DataFrame(results, columns=['city', 'seg_id'])

# Step 4: 构建城市网络
G = nx.Graph()
for city_name in city['cityname_eng']:
    G.add_node(city_name)

# 同一铁路段上的城市两两相连
for seg_id, group in df_link.groupby('seg_id'):
    cities = group['city'].tolist()
    for i in range(len(cities)):
        for j in range(i+1, len(cities)):
            G.add_edge(cities[i], cities[j], line=seg_id)

print("Total cities:", G.number_of_nodes())
print("Total edges:", G.number_of_edges())

import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Point

# 城市点
city_gdf = gpd.GeoDataFrame(city, geometry=gpd.points_from_xy(city.coord_long, city.coord_lat), crs="EPSG:4326")

# 可选：从铁路像素生成铁路线段
rail_lines = []
valid_seg_ids = []

for seg, group in pixel.groupby('seg_id'):
    coords = list(zip(group['long'], group['lat']))
    if len(coords) >= 2:  # ✅ 至少两个点才画线
        rail_lines.append(LineString(coords))
        valid_seg_ids.append(seg)

rail_gdf = gpd.GeoDataFrame({'seg_id': valid_seg_ids}, geometry=rail_lines, crs="EPSG:4326")


# 画图
fig, ax = plt.subplots(figsize=(12, 10))
rail_gdf.plot(ax=ax, color='black', linewidth=0.7, alpha=0.5)
city_gdf.plot(ax=ax, color='red', markersize=20)

# 画城市间连线
for u, v in G.edges():
    p1 = city_gdf.loc[city_gdf['cityname_eng']==u, 'geometry'].values[0]
    p2 = city_gdf.loc[city_gdf['cityname_eng']==v, 'geometry'].values[0]
    line = LineString([p1, p2])
    gpd.GeoSeries([line], crs="EPSG:4326").plot(ax=ax, color='red', alpha=0.5)

plt.title("China Railway City Network (pixel-based connection)")
plt.show()
```

```python
import pandas as pd
import numpy as np
import networkx as nx
from scipy.spatial import cKDTree

city = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

# 建立像素索引
coords = pixel[['pos_x', 'pos_y']].values
tree = cKDTree(coords)

# 构建铁路像素连通图
G_pixels = nx.Graph()
for i, (x, y) in enumerate(coords):
    neighbors = tree.query_ball_point([x, y], r=1.5)  # 8邻域像素
    for j in neighbors:
        if i != j:
            G_pixels.add_edge(i, j)

# 找到连通分量（每个铁路子网络）
components = list(nx.connected_components(G_pixels))
pixel['component'] = -1
for cid, comp in enumerate(components):
    pixel.loc[list(comp), 'component'] = cid

# 匹配城市到铁路分量
city_to_comp = []
for _, c in city.iterrows():
    dist, idx = tree.query([c['pos_x'], c['pos_y']])
    if dist < 3:
        comp = pixel.iloc[idx]['component']
        city_to_comp.append((c['cityname_eng'], comp))

df_comp = pd.DataFrame(city_to_comp, columns=['city', 'component'])

# 构建城市连通网络
G_city = nx.Graph()
for cid, group in df_comp.groupby('component'):
    cities = group['city'].tolist()
    for i in range(len(cities)):
        for j in range(i + 1, len(cities)):
            G_city.add_edge(cities[i], cities[j])

print("City network constructed!")
print("Nodes:", G_city.number_of_nodes(), "Edges:", G_city.number_of_edges())

import geopandas as gpd
from shapely.geometry import Point, LineString
import matplotlib.pyplot as plt

# 城市点
gdf_city = gpd.GeoDataFrame(city, geometry=gpd.points_from_xy(city.coord_long, city.coord_lat), crs="EPSG:4326")

# 绘制
fig, ax = plt.subplots(figsize=(10, 8))
gdf_city.plot(ax=ax, color='red', markersize=30)

# 城市间边
for u, v in G_city.edges():
    p1 = gdf_city.loc[gdf_city['cityname_eng']==u, 'geometry'].values[0]
    p2 = gdf_city.loc[gdf_city['cityname_eng']==v, 'geometry'].values[0]
    gpd.GeoSeries([LineString([p1, p2])], crs="EPSG:4326").plot(ax=ax, color='gray', linewidth=0.5, alpha=0.7)

plt.title("China Railway Network (Pixel Connected City Graph)")
plt.show()
```

```python
import pandas as pd
import networkx as nx
from sklearn.neighbors import BallTree
import numpy as np

city = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pref_pair/cityinfo.csv")
pixel = pd.read_csv("/Users/tshenyi/UMich/2025fall/si608/transportation_networks_of_china/pixel_info/pixel_info_rail_good_2017.csv")
pixel.columns = pixel.columns.str.strip()

coords_pixel = np.radians(pixel[['lat', 'long']])
coords_city = np.radians(city[['coord_lat', 'coord_long']])

tree = BallTree(coords_pixel, metric='haversine')
dist, idx = tree.query(coords_city, k=1)

city['nearest_seg'] = pixel.iloc[idx.flatten()]['seg_id'].values

G = nx.Graph()

# 添加城市节点
for c in city['cityname_eng']:
    G.add_node(c)

# 按 seg_id 连接城市
for seg, group in city.groupby('nearest_seg'):
    cities = group['cityname_eng'].tolist()
    for i in range(len(cities)):
        for j in range(i + 1, len(cities)):
            G.add_edge(cities[i], cities[j], seg_id=seg)

import matplotlib.pyplot as plt

pos = {row.city: (row.lon, row.lat) for _, row in city.iterrows()}
nx.draw(G, pos, node_size=8, font_size=5, width=0.3)
plt.title("China Railway Network (City-level)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

```
