*[[5 points]]*

Load `facebook_combined.txt` into networkx. This is an ego-network of Facebook connections retrieved from http://snap.stanford.edu/data/ego-Facebook.html. Friendships on facebook are mutual; therefore all edges here are undirected.

````python
# YOUR CODE HERE
import networkx as nx
edges = []
with open('facebook_combined.txt', 'r') as f:
    for line in f:
        u, v = map(int, line.strip().split(' '))
        edges.append((u, v))

G = nx.Graph()
G.add_edges_from(edges)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")

# average degree
# the sum of degree
avg_degree = sum(dict(G.degree).values())/G.number_of_nodes()
print(f"Average degree: {avg_degree:.4f}")

# average clustering
avg_clus = nx.average_clustering(G)
print(f"Average clustering: {avg_clus:.4f}")
````

### **Q5.2 |**  Calculate clustering coefficients using NetworkX 
*[[5 points]]*
    
In class, we introduced (local) clustering coefficient as "the probability that two randomly selected friends of a node are friends with each other."


Here's how `NetworkX` operationalizes this...

> For unweighted graphs, the clustering of a node $u$ is the fraction of possible triangles through that node that exist, $c_u = \frac{2T(u)}{deg(u)(deg(u)-1)}$,   
where $T(u)$ is the number of triangles through node $u$ and $deg(u)$ is the degree of $u$.
-- [networkx documentation](https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.cluster.clustering.html#networkx.algorithms.cluster.clustering)

`nx.clustering(g)` will return a full dictionary of all clustering coefficients.   
`nx.clustering(g, node)` will return the corresponding clustering coefficient for a specific node.

Now plot the distribution of clustering coefficients for this network.   
*Hint: use `plt.hist()` to generate a histogram.*

````python
# YOUR CODE HERE
import matplotlib.pyplot as plt

# draw all the clustering coefficients
clus_coef = list(dict(nx.clustering(G)).values())
plt.hist(clus_coef)
plt.title("The distribution of Clustering Coefficient")
plt.xlabel("Clustering Coefficient")
plt.ylabel("Frequency")
plt.show()
````

### **Q5.3 |**  Build your own function to calculate clustering coefficients 
*[[8 points]]*

Let's now break down how `nx.clustering()` computes the score. The only `NetworkX` function we'll use is `g[node]` (to retrieve the node's neighbor list). 

````python
def node_clustering_coeff(g, n):
    # This function should return a clustering coefficient for node n in graph g
    
    # STEPS:
    # 1. get degree of node n
    # 2. count triangles for node n
    # 3. compute clustering coefficient
    
    # STEP 1: node degree
    neighbors = set(dict(g[n]).keys()) # YOUR CODE HERE  # collect neighbors for node n (remember to exclude n itself in case there are self-loop links)
    degree = len(neighbors)# YOUR CODE HERE
    
    # STEP 2: number of triangles
    n_triangles = 0
    for nbr in neighbors:  # interate through each neighbor
        neighbor_nbr = set(dict(g[nbr]).keys())# YOUR CODE HERE  # get neighbors for this neighbor (again, exclude n)
        n_common_nbr = len(neighbors.intersection(neighbor_nbr))# YOUR CODE HERE  # how many of their neighbors is also node n's neighbor?
        # YOUR CODE HERE  # for each set of common neighbors, there's a triangle (double counted)
        n_triangles += n_common_nbr
    
    # STEP 3: Now compute the clustering coefficient given the triangles you found
    # YOUR CODE HERE
    if degree == 0 or degree == 1:
        clustering_coeff = 0
    else:
        clustering_coeff = 2*n_triangles/(degree*(degree-1))
    
    return clustering_coeff
````

Once you have that, calculate the pearson or spearman correlation between the status field and the  centrality measures you computed (you will have 5 correlation measures).
