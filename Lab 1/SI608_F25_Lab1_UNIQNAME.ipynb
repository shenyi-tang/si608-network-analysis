**Now draw a histogram of the degree distribution. Do this manually without writing code:**


Here’s a program to draw the degree distribution:
```
degree = g.degree()    
# a list of all the degree values: 
degreevals = [val for (node, val) in degree]
degvaluniq = sorted(set(degreevals))

# this gets all the unique degree values, and sorts them
# so degvaluniq = [0, 1, 2, 3]
```

Now we want to know how many nodes are of each degree (how many have degree 0, 1, 2, 3, 4, etc.)
```
hist = [degreevals.count(x) for x in degvaluniq]   
# which finds the number of nodes that have each degree count
```

Now, lets draw it:
```
plt.figure()
plt.plot(degvaluniq,hist,'ro-')
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
plt.show()

```

And here is another option to draw the degree distribution:
```
degree = g.degree()    
degreevals = [val for (node, val) in degree]
plt.hist(degreevals,bins=range(0,5), rwidth=1,align='left') #use matplotlib histogram function
plt.xlabel("Degree")
plt.ylabel("Number of nodes")
```
